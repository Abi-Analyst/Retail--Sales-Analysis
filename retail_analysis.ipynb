```python
#Real Data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Data/amazon_sales_dataset.csv")
```

```python
#data cleaning
df.columns = df.columns.str.strip()
df.dropna(inplace = True)
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
#feature engineering
df['YearMonth'] = df['order_date'].dt.to_period('M')
df['Month'] = df['order_date'].dt.month
df['Year'] = df['order_date'].dt.year
```

```python
#Business KPI engine
#Revenue KPIs
total_revenue = df['total_sales'].sum()
average_order = df['total_sales'].mean()
median_sales = df['total_sales'].median()
print('Total Revenue:',total_revenue)
print('Average Order:', average_order)
print('Median sales:', median_sales)
```

```python
#Numpy Statistical engine
sales_array = df['total_sales'].values
print("Standard Deviation:",np.std(sales_array))
print("90th Percentile:",np.percentile(sales_array,90))
print("Correlation Shipping vs Sales:",np.corrcoef(df['shipping_cost'],sales_array))
```

```python
#Advanced data analytics
#Fraud and detection logic
#Detecting suspicious transactions
high_discount = df['discount']>0.25
high_value = df['total_sales']>100000

fraud_cases = df[high_discount & high_value]
print("Potential Fraud Cases:",len(fraud_cases))
```

```python
#Customer Segmentation
df['customer_segment']=df['total_sales'].apply(customer_segment)
```

```python
#visual analytical dashboard
monthly_sales = df.groupby('YearMonth')['total_sales'].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index.astype(str),monthly_sales.values)
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
#category performance
category_sales = df.groupby('category')['total_sales'].sum()
category_sales.plot(kind="bar")
plt.title("Category Revenue")
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
#discount impact
plt.scatter(df['discount'],df['total_sales'],alpha = 0.3)
plt.title("Discount vs Sales")
plt.show()
```

```python
#Iterators
sales_iter = iter(sales_array)
print(next(sales_iter))
print(next(sales_iter))
```

```python
#string operators
df['Customer Name length']=df['customer_name'].str.len()
```

```python
#range + advanced looping
for year in range(df['Year'].min(),df['Year'].max() + 1):
    yearly_sales = df[df['Year'] == year]['total_sales'].sum()
    print(f"Year{year} Sales:", yearly_sales)
```

```python
#Final business insights function
def generate_business_report(df):
    print("==== BUSINESS REPORT ====")
    print("Total Revenue:",df['total_sales'].sum())
    print("Top Category:",df.groupby('category')['total_sales'].sum().idxmax())
    print("Top Payment Method:",df['payment_method'].mode()[0])
    print("High Value Transactions:",len(df[df['total_sales']>100000]))
generate_business_report(df)   
```

```python
#total spend of customers
customer_spend= df.groupby('customer_id')['total_sales'].sum().reset_index()
customer_spend.head()
```

```python
#customer segmentation logic(nested if)
def segment_customer(amount):
    if amount>200000:
        return "VIP"
    elif amount> 100000:
        return "premium"
    elif amount>50000:
        return "Regular"
    else:
        return "Low value"
```

```python
#apply segmentation
customer_spend['segment']=customer_spend['total_sales'].apply(segment_customer)
customer_spend.head()
```

```python
class Order:
    
    def __init__(self, order_id, customer_id, category, total_sales):
        self.order_id = order_id
        self.customer_id = customer_id
        self.category = category
        self.total_sales = total_sales
    
    def calculate_discount(self):
        if self.total_sales > 200000:
            return 0.20
        elif self.total_sales > 100000:
            return 0.15
        elif self.total_sales > 50000:
            return 0.10
        else:
            return 0.05
    
    def final_amount(self):
        discount = self.calculate_discount()
        return self.total_sales * (1 - discount)
    
    def order_segment(self):
        if self.total_sales > 200000:
            return "High Value"
        elif self.total_sales > 100000:
            return "Medium Value"
        else:
            return "Standard"
```

```python
#create an object
order1= Order(101,5001,'Electronics',150000)
print('Discount:',order1.calculate_discount())
print('Final Amount:',order1.final_amount())
print('segment:',order1.order_segment())
```

```python
#apply oop to real dataset
orders = []

for index, row in df.head(5).iterrows():
    order = Order(
        row['order_id'],
        row['customer_id'],
        row['category'],
        row['total_sales']
    )
    orders.append(order)
```

```python
for o in orders:
    print(o.order_id, o.order_segment(),o.final_amount())
```
