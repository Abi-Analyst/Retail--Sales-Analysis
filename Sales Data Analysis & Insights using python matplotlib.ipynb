```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#To display plots properly
%matplotlib inline
```

```python
df=pd.read_csv("amazon_sales_dataset.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
#DATA CLEANING
df.isnull().sum()
```

```python
df.dropna()
```

```python
#remove duplicates
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
#convert date column
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
#create new columns
df['Year'] = df['order_date'].dt.year
df['Month'] = df['order_date'].dt.month
```

```python
#Exploratory data analysis(eda)
#sales by region
region_sales = df.groupby('city')['total_sales'].sum().sort_values(ascending = False)

top10 = region_sales.head(10)
top10.plot(kind='bar',figsize=(10,5))
plt.title("Top 10 cities by Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
monthly_sales = df.groupby('Month')['total_sales'].sum().sort_index()

monthly_sales.plot(kind ='line', figsize = (10,5))
plt.title("Monthly sales Trend")
plt.ylabel = ("Total sales")
plt.ticklabel_format(style='plain',axis='y')
plt.xticks(range(1,13))
plt.show()
```

```python
#which category generated the highnest revenue
category_sales = df.groupby('category')['total_sales'].sum().sort_values(ascending = False)

category_sales.plot(kind='bar',figsize=(8,5))
plt.title("sales by category")
plt.ylabel=("Total sales")
plt.ticklabel_format(style='plain',axis='y')
plt.xticks(rotation=45)
plt.show()
```

```python
#Which specific subcategory performs best?
sub_category = df.groupby('sub_category')['total_sales'].sum().sort_values(ascending=False)

top10_sub = sub_category.head(10)

top10_sub.plot(kind='barh',figsize = (8,6))
plt.title("Top 10 Sub category by sales")
plt.ylabel=('Total Sales')
plt.ticklabel_format(style='plain',axis='x')
plt.gca().invert_yaxis()
plt.show()
            
```

```python
#which products bring  highest revenue?
product_sales = df.groupby('product_name')['total_sales'].sum().sort_values(ascending = False)

top10_products= product_sales.head(10)

top10_products.plot(kind='barh',figsize = (8,6))
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ticklabel_format(style='plain',axis='x')
plt.gca().invert_yaxis()
plt.show()
```

```python
#which state generates highest revenue?
state_sales = df.groupby('state')['total_sales'].sum().sort_values(ascending = False)

top10_states= state_sales.head(10)

top10_states.plot(kind='bar',figsize = (10,5))
plt.title("Top 10 States by Sales")
plt.xlabel("Total Sales")
plt.ticklabel_format(style='plain',axis='y')
plt.xticks(rotation=45)
plt.show()
```

```python
#which payment method is most used?
payment_sales = df.groupby('payment_method')['total_sales'].sum()
payment_sales.plot(kind='pie',autopct='%1.1f%%',figsize=(6,6))
plt.title('Sales Distribution by Payment Method')

plt.show()
```

```python
#does higher discount increase sales?
plt.figure(figsize=(8,5))
plt.scatter(df['discount'],df['total_sales'],alpha=0.3)
plt.title("Discount vs Total Sales")
plt.xlabel("Discount")
plt.ylabel="Total sales"
plt.ticklabel_format(style='plain',axis='x')
plt.show()
```

```python
discount_analysis = df.groupby('discount')['total_sales'].mean()

discount_analysis.plot(kind='line', marker='o', figsize=(8,5))

plt.title("Average Sales by Discount")
plt.xlabel("Discount")
plt.ylabel="Average Total Sales"
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
#how many orders are delivered vs cancelled?
status_counts = df['order_status'].value_counts()

status_counts.plot(kind='bar')
plt.title("Order status Distribution")
plt.ylabel="Number of Orders"
plt.show()
```

```python
#Customer Analysis
#top customers by revenue
customer_sales = df.groupby('customer_name')['total_sales'].sum().sort_values(ascending=False)

top10_customers = customer_sales.head(10)

top10_customers.plot(kind='barh',figsize=(8,6))
plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Shipping cost Analysis
#Does shipping cost impact total sales?
plt.scatter(df['shipping_cost'],df['total_sales'])
plt.title("Shipping Cost vs Total sales")
plt.xlabel("Shipping Cost")
plt.ylabel="Total Sales"
plt.show()
            
```

```python
#using hexbin
plt.figure(figsize=(8,5))
plt.hexbin(df['shipping_cost'],df['total_sales'],gridsize=30)

plt.title("Shipping Cost vs Total sales(density)")
plt.xlabel("Shipping cost")
plt.ylabel="Total Sales"

plt.colorbar(label="count")
plt.ticklabel_format(style='plain',axis='y')
plt.show()

           
```

```python
#using Trendline
import numpy as np
plt.figure(figsize=(8,5))
plt.scatter(df['shipping_cost'],df['total_sales'],alpha=0.2)
#Adding regression line
m,b = np.polyfit(df['shipping_cost'],df['total_sales'],1)
plt.plot(df['shipping_cost'],m*df['shipping_cost']+b,color='red')

plt.title("Shipping Cost vs Total Sales")
plt.xlabel("Shipping Cost")
plt.ylabel='Total Sales'

plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
#Quantity Analysis
#Which products are sold most(by quantity)?
quantity_analysis = df.groupby('product_name')['quantity'].sum().sort_values(ascending=False)

top10_quantity=quantity_analysis.head(10)

top10_quantity.plot(kind='barh',figsize=(8,6))

plt.title("Top 10 Products by quantity sold")
plt.xlabel("Total Quantity")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Line Chart
#Monthly sales Trend
df['order_date']=pd.to_datetime(df['order_date'])
df['YearMonth']=df['order_date'].dt.to_period('M')

monthly_sales=df.groupby('YearMonth')['total_sales'].sum()
plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index.astype(str),monthly_sales.values)

plt.title("Monthly Sales Trend")
plt.xlabel("Year-Month")
plt.ylabel="Total Sales"
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
#Histogram (Distribution)
#Distribution of Total sales
plt.figure(figsize=(8,5))
plt.hist(df['total_sales'],bins=30)

plt.title('Distribution of Total Sales')
plt.xlabel("Total Sales")
plt.ylabel='Frequency'
plt.ticklabel_format(style='plain',axis='x')
plt.show()
```

```python
#Box Plot(Outlier Detection)
#Sales by Category
plt.figure(figsize=(8,5))
df.boxplot(column="total_sales",by='category')

plt.title("Sales Distribution by category")
plt.suptitle("")
plt.xlabel("Category")
plt.ylabel="Total Sales"
plt.show()
```

```python
#Stacked bar chart
#Sales by category & payment method
stack_data=df.pivot_table(values='total_sales',index='category',columns='payment_method', aggfunc= 'sum')

stack_data.plot(kind='bar',stacked=True,figsize=(8,5))
    
plt.title("Category Sales by payment method")
plt.ylabel='Total Sales'
plt.xticks(rotation=45)
plt.show()
```

```python
#Area Chart
monthly_sales.plot(kind='area',figsize=(10,5))
plt.title("Monthly sales Area chart")
plt.ylable='Total sales'
plt.show()
```

```python
#heatmap(correlation)
import seaborn as sns
plt.figure(figsize=(6,4))
sns.heatmap(df[['quantity','unit_price','discount','shipping_cost','total_sales']].corr(),annot=True)

plt.title("Correlation Heatmap")
plt.show()
```
